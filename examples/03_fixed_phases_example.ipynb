# **03 Model ∆G of fixed phases**
---

This notebook shows how to:
1. Add fixed phases to a Theriak database
2. Run a minimisation
3. Read the delta G values
4. Append or overwrite the fixed phases

The example removes the updated database copy at the end so the workspace stays clean.

---

### Import all required libraries.

```python
from pathlib import Path
from pytheriak import wrapper
```

### Initiate a TherCaller-object.
The argument `programs_dir` must be set to the directory of your Theriak-Domino installation.\
The `database`-file (`"td-ds62-mp50-v07.txt"`) must be in your working directory, alongside this jupyter notebook. Get this database from [Databases](https://github.com/Theriak-Domino/Databases) on github.\
The Theriak-Domino database `"td-ds62-mp50-v05.txt"` is based on the Holland and Powell (2011, https://doi.org/10.1111/j.1525-1314.2010.00923.x) updated thermodynamic dataset file with accompanying solution models for metapelites.

```python
theriak = wrapper.TherCaller(
    programs_dir=Path.home() / "TheriakDominoMAC" / "Programs",
    database="td-ds62-mp50-v05.txt",
    theriak_version="v2023.06.11",
    verbose=True,
)

bulk = "SI(68.2)TI(0.76)AL(25.18)FE(9.96)MN(0.02)MG(4.36)CA(0.18)NA(0.06)K(7.74)H(100)O(?)O(0.01)"
```

### Append some fixed phases

The first call appends two fixed garnet compositions to a copied database and runs Theriak once.

```python
updated_db = theriak.append_fixed_phases(
    names=["Garnet_core", "Garnet_rim"],
    solnam_list=["GRT", "GRT"],
    em_frac_list=[
        "py(0.03) alm(0.40) gr(0.35) spss(0.21)",
        "py(0.05) alm(0.45) gr(0.30) spss(0.20)",
    ]
)

rock, element_list = theriak.minimisation(
    pressure=5000,
    temperature=550,
    bulk=bulk,
)

print("Append example")
print("Garnet_core deltaG:", rock.mineral_delta_G["Garnet_core"])
print("Garnet_rim deltaG:", rock.mineral_delta_G["Garnet_rim"])
```

### Overwrite fixed phases

The second call overwrites the fixed-phase block with new compositions, then runs Theriak again.

```python
theriak.overwrite_fixed_phases(
    names=["Garnet_core", "Garnet_rim"],
    solnam_list=["GRT", "GRT"],
    em_frac_list=[
        "py(0.04) alm(0.42) gr(0.32) spss(0.22)",
        "py(0.06) alm(0.44) gr(0.28) spss(0.22)",
    ]
)

rock, element_list = theriak.minimisation(
    pressure=5000,
    temperature=450,
    bulk=bulk,
)

print("Overwrite example")
print("Garnet_core deltaG:", rock.mineral_delta_G["Garnet_core"])
print("Garnet_rim deltaG:", rock.mineral_delta_G["Garnet_rim"])
```

```python
# clean up the updated database file
updated_db.unlink()
```
